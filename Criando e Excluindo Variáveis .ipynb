## Criando Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head()
```

## Criando variáveis Valor Bruto, Valor por m2 , Valor Bruto/m2 e Tipo Agregado

### Criando Valor Bruto

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

### Criando Valor por m2

```python
dados['Valor por m2'] = (dados['Valor']/ dados['Area']).round(2)
dados.head()
```

### Criando Valor Bruto por m2

```python
dados['Valor Bruto por m2'] = (dados['Valor Bruto']/ dados['Area']).round(2)
dados.head()
```

### Criando Tipo Agregado

```python
dados['Tipo'].drop_duplicates()
```

```python
casa = ['Quitinete', 'Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
