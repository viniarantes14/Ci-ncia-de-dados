### Imoveis sem valor

```python
quantidade_de_imoveis_sem_valor = 32960 - 32943 
print('A base de Dados aprensenta {} imoveis sem valor'.format(quantidade_de_imoveis_sem_valor))
```

```python
quantidade_de_imoveis_sem_condominio = 32960 -  28867
print('A base de Dados aprensenta {} imoveis sem condominio'.format(quantidade_de_imoveis_sem_condominio))
```
