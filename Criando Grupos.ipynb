## Criando Grupos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head()
```

```python
dados['Bairro'].unique()
```

```python
grupo_bairro = dados.groupby('Bairro')
grupo_bairro.groups
```

```python
type(grupo_bairro)
```

## Selecionando a Média de Valores pelo laço for

```python
for bairro, data in grupo_bairro: 
    print('{} -> {}'.format(bairro, data['Valor'].mean()))
```

## Selcionando Valores Direto

```python
(grupo_bairro[['Valor', 'Condominio']].mean()).round(2)
```

## Estatísticas Descritivas

### Utilizando Describe()

```python
grupo_bairro['Valor'].describe().round(2)
```

### Utilizando Aggregate()

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'std']).rename(columns = {'min': 'Mínimo', 
                                                                         'max': 'Máximo', 
                                                                         'std': 'Desvio Padrão'})
```

### Como visto anteriormente para uma base de dados grande é inviavel olhar 1 por um , para isso a  solução utilizada será a geração de gráficos para realizar essa análise

## Gerando gráficos

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### Configurando tamanho dos graficos

```python
plt.rc('figure', figsize = (200, 90))
plt.rcParams['xtick.labelsize'] =100
plt.rcParams['ytick.labelsize'] = 150
```

### Plotando e configurando Gráficos de Barra 

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize': 180})
fig.set_title('Desvio Padrão por Bairro', {'fontsize': 180})
```
